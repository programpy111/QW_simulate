```python
import numpy as np
import matplotlib.pyplot as plt

n = 7200
k = 1000
x = np.zeros((1,k))
y = np.zeros((1,k))
x[0][500] = 100
x[0][501] = 100

for i in range(n):
    y[0][k-1] = float(x[0][k-2]/2)
    y[0][0] = float(x[0][1]/2)
    for j in range(k-3):
        y[0][j+1] = float(x[0][j]/2) + float(x[0][j+2]/2)

    x = np.copy(y)


# x軸の値を生成（例えば 0 から k-1 までの整数）
x_values = np.arange(k)

# プロット作成（yの最終状態をプロット）
plt.plot(x_values, x[0], label='Modified Values')

# タイトルとラベルを追加
plt.title('random walk')
#plt.xlabel('X axis')
#plt.ylabel('Y axis')

# 凡例を表示
#plt.legend()

# グラフを表示
plt.show()
```
